```python
def uniplot(af,col,title,hue =None):
    fig = plt.figure(figsize=[20,15])
    plt.rcParams["axes.labelsize"] = 25
    plt.rcParams['axes.titlesize'] = 25
    plt.rcParams['axes.titlepad'] = 18
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(appdata[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 10)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = af , x= col, order=af[col].value_counts().index,hue = hue, palette = 'hls') 
        
# PLotting for income range (AMT_INCOME_RANGE)

uniplot(target0_ad,col='AMT_INCOME_RANGE',title='Distribution of income range', hue='CODE_GENDER')
plt.show()
```

```python
# Plotting the above correlation with heatmap, this will be the best choice to visulaise the same

def targets_corr(data,title):
    plt.figure(figsize=(32, 32))
    plt.rcParams['axes.titlesize'] = 25
    plt.rcParams['axes.titlepad'] = 65

    sns.heatmap(data, cmap="RdYlBu",annot=True)

    plt.title(title)
    plt.yticks(rotation=0)
    plt.show()
```

```python
# Heatmap plot of Target 0

targets_corr(data = target0, title ='Correlation of target 0')
```

### Inference

##### From the above correlation heatmap, we can inference below points:

- Income amount is inversely proportional to the number of children client have, means more income for less children client have and vice-versa.
- Credit amount is inversely proportional to the date of birth, that means Credit amount is higher for low age and vice-versa.
- Credit amount is inversely proportional to the number of children client have, that means Credit amount is higher for less children count client have and vice-versa.

- In densely populated area:
    - Higher income
    - Having less children Clients
    - Credit Amount is higher
